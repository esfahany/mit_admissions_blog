MIT Students: During this school year, at approximately what time have you typically woken up on weekdays? 


```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['pdf.fonttype'] = 42 # settings to make figure text editable in Adobe Illustrator
plt.rcParams['ps.fonttype'] = 42
```

```python
# labels taken from the survey
wake_times = ["≤ 5 AM", "6 AM", "7 AM", "8 AM", "9 AM", "10 AM", "11 AM", "≥ 12 PM"]
sleep_time_labels = ["≤ 10 PM", "11 PM", "12 AM", "1 AM", "2 AM", "3 AM", "≥ 4 AM"]
```

```python
N = len(wake_times) # number of categories for wake-up time
```

```python
# wake data by sleep time
# explanation of values: in _10_PM, 0.9 is people who sleep at 10 and wake up before 5 
_10_PM = [0.9, 2.7, 1.7, 0.5, 0.0, 0.0, 0.0, 0.0]
_11_PM = [0.8, 3.2, 9.1, 4.2, 1.1, 0.1, 0.0, 0.0]
_12_AM = [0.3, 1.7, 7.4, 10.7, 3.2, 0.6, 0.1, 0.0]
_1_AM = [0.1, 0.5, 2.7, 8.5, 7.4, 1.9, 0.4, 0.1]
_2_AM = [0.1, 0.4, 1.0, 3.9, 6.3, 5.2, 1.0, 0.2]
_3_AM = [0.0, 0.0, 0.3, 1.1, 2.1, 2.3, 1.7, 0.6]
_4_AM = [0.0, 0.0, 0.0, 0.4, 0.5, 1.0, 0.6, 0.9]

sleep_times = [_10_PM, _11_PM, _12_AM, _1_AM, _2_AM, _3_AM, _4_AM]
```

```python
# calculate the total percent who wake up at each time

wake_total = []
for j in range(N): # for each wake-up time
    s = 0
    for i in range(len(sleep_times)): # add up the sleep time subcategories
        s += sleep_times[i][j]
    wake_total.append(np.round(s,2))
print(wake_total)
```

```python
# simple plot of people by their wake up time
plt.figure(figsize = (4,3), dpi = 200)
plt.bar(np.arange(N), wake_total)
plt.xticks(np.arange(N), wake_times, rotation = 45)
plt.xlabel("Wake Up Time")
plt.ylabel("Percentage of Respondents")
plt.show()
```

```python
# stacked bar plot

fig = plt.figure(figsize = (4,3), dpi = 400)

colors = ["#9A2006", "#DF3F2D", "#DF682D", "#EB8B29", "#EBA429", "#F6C223", "#FEDC27"] # red, orange, yellow like a sunrise

running_total = np.zeros(N)

for i in range(len(sleep_times)): # for each sleep time group
    plt.bar(np.arange(N), sleep_times[i], bottom = running_total, color = colors[i])
    running_total += sleep_times[i]
    
plt.xticks(np.arange(N), wake_times, rotation = 45)
plt.xlabel("Wake Up Time")
plt.ylabel("Percentage of Respondents")
lgd = plt.legend(sleep_time_labels, title = "sleep time", bbox_to_anchor=(1, 1), framealpha = 0.0)
plt.title("Sleeping Habits at MIT")
# fig.savefig('sleeping_habits.pdf', bbox_extra_artists=(lgd,), bbox_inches='tight')
plt.show()
```
